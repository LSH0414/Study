# 예측 될 수 있는 것?

- 예측의 요인
  - 영향을 주는 요인을 얼마나 잘 이해할 수 있는가?
  - 사용할 수 있는 데이터의 량은 얼마나 되는가?
  - 예측이 우리ㅏ 예측하려는 것에 영향을 줄 수 있는가?

### 예측
예측이란 주어진 이용가능한 모든 정보(과거 데이터 및 예측에 영향을 끼칠 수 있는 미래 사건에 대한 지식을 포함)를 바탕으로 가능한 한 정확하게 미래를 예측하는 것.

- 예측의 종류
  - 단기</br>
  > 인사, 생산, 수송, 계획 등에 필요. 계획 과정의 한 부분으로 수요 예측도 필요.
  - 중기</br>
  >원자재 구입, 신규 채용, 장비나 기계 구입 등 미래 자원 공급을 결정하는데 사용
  - 장기</br>
  >전략적으로 계획을 세우는데 사용. 시장 기회, 환경 요인, 내부 자원을 반드시 고려하여 결정

</br></br>

### 목표
목표는 예측 및 계획과 연관되어 있는 것이 좋지만, 항상 일어나는 것은 아니다. 목표를 어떻게 달성할지에 대한 계획과 목표가 실현 가능한지에 대한 예측 없이 목표를 세우는 것은 좋지 않다.


</br></br>

### 계획
계획은 예측과 목표에 대한 대응이다. 계획은 예측과 목표를 일치시키는데 필요한 적절한 행동을 결정하는 일을 포함한다. 



### 예측 데이터와 기법

이용할 수 있는 데이터가 없거나, 이용할 수 있는 있지만 예측에 상관 없는 데이터라면 **정성적인 예측 기법**을 사용

**정량적인 예측**은 두 가지 조건을 만족할 때 사용.
- 과거 수치 정보를 사용할 수 있을 때
- 과거 패턴의 몇 가지 양상이 미래에도 계속될 것이라고 가정하는 것이 합리적일 때


#### 예측 변수와 시계열 예측

시간당 전기 수요(ED)를 예측한다고 가정한다면
$${ED = f(현재기온, 경제상황, 인구, 시간, 요일, 오차}$$

예측 변수로 완벽하게 설명할 수 없는 전기 수요량은 항상 변화할 것이기에 이 관계는 정확하지 않다. 우변에 있는 '오차'항은 무작위 변동과 더불어 모델에 포함하지 않은 관련 변수의 영향까지 고려하는 역할을 한다. 이는 어떤 것이 전기 수요의 변동을 일으키는지 도움이 되기 때문에 **설명 모델**이라고 한다.

수요 데이터의 시계열 예측식은 다음과 같다.
$${ED_{t+1} = f(ED_t, ED_{t-1}, ED_{t-2}..., 오차)}$$

t는 현재시간 t+1은 한 시간 후, t-1은 한 시간 전을 나타낸다. 변수의 과거 값으로 미래 값을 예측한다.

위 두 가지 모델을 결합한 모델이 있다.
$${ED_{t+1} = f(ED_t,현재 기온, 시간, 요일, 오차)}$$

이러한 종류의 혼합된 모델은 분야에 따라 다양한 이름으로 불린다. 동적 회귀 모델, 패널 데이터 모델, 종단적 모델, 수송 함수 모델, 선형 시스템 모델

**시계열 모델을 선택하는 경우**

- 시스템을 잘 이해하지 못하거나, 이해하더라도 변수의 행동에 영향을 주는 관계를 측정하기 아주 어려운 경우
- 관심 있는 변수를 예측할 수 있으려면 다양한 예측 변수의 미래값을 알 필요가 있거나 예측할 필요가 있는 경우
- 주된 관심이 왜 일어나는지가 아니라 무엇이 일어나는지만 예측하는데 있는 경우가 있다.
- 시계열 모델이 설명 모델이나 혼합 모델보다 성능이 더 좋을 경우이다. 

사용할 수 있는 자원과 데이터, 정확도, 사용 방식에 따라 사용할 모델이 달라진다.

## 예측 작업의 기본 단계

- **1단계 : 문제 정의**
> 예측값이 어떤 방식으로 사용될지, 누가 예측값을 요구하는지, 예측 함수가 예측값을 요구하는 조직에 얼마나 잘 어울릴지를 모두 이해할 필요가 있다.

- **2단계 : 정보 수집**
> 최소한으로 요구되는 데이터에는 두 종류가 있다.
(1) 통계 데이터, (2) 데이터를 모으고 예츠값을 사용하는 사람들에게 축적된 전문지식
통계 데이터가 부족할 경우 판단예측기법을 사용할 수도 있다. 충분한 데이터가 있을때 구조적 변화로 과거의 데이터를 사용하지 않고 최근의 데이터만 사용할 수 있지만 좋은 모델은 이러한 변화에도 대응할 수 있는 모델이다.

- **3단계 : 예비분석**
> **데이터는 항상 그래프를 나타내는 것으로 시작하자** 패턴, 추세, 계절성, 순환, 전문적 지식으로 설명 가능한 특이점이 있는지 확인하자.

- **4단계 : 모델 선택**
> 과거 데이터를 이용할 수 있는 정도, 예측변수와 설명 변수 사이 관계의 강도, 그리고 예측값이 사용될 방식에 따라 사용할 가장 좋은 모델이 달라진다.

- **5단계 : 예측 모델 사용 그리고 평가**
> 모델을 선택하고, 매개변수를 추정한 후 예측값을 내는 요도로 모델을 사용한다.
