# 03. 자기상관


## 자기상관(Autocorrelation)
.  자기상관(Autocorrelation)
: 경제시계열의 경우 과거가 현재, 현재가 미래에 영향을 주어 추세나 순환현상을 보이는 경우가 많다. 이러한 현상을 보이는 이뉴는 경제에 외부 충격(예상치 못한 변화)이 주어졌을때 그 충격의 영향이 단기간에 끝나지 않고 장기적으로 지속되기 떄문이다. 시계열에 있어서 t의 교란값과 t-1의 교란값이 상관관계가 있을 경우 계열상관(Serialcorrelation) 혹은 자기상관(Autocorrelation)이 있다고 한다. 이 경우를 1차 지기상관이라한다.

$${Y_t} = {X_tB + u_t}$$
$${u_t} = {pu_{t-1}+e_t}$$


시계열의 자기상관 여부를 탐지하는 방법으로는
- 그래프 분석 : 회귀식으로부터 도출된 자차를 그려 보아 자기상관 여부를 판단하는 방법

- 통계적 검정 Durbin-Watson 검정, LM(Lagrange Multiplier) 검정이 있다.



## 그래프 분석
: 자기상관을 탐지하는 가장 간단한 방법으로 잔차에 상관관계가 나타나는 지를 그래프를 통해 살펴보는 것인데 잔차가 아래와 같은 경향을 보인다면 자기상관이 있는 것으로 판단하는데 이 방법은 주관적이며, 과학적이지 못하다.

## Durbin_Watson Test
: 잔차의 자기상관 유무를 검정하는 기법

![img](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FxfuSm%2FbtqM9PKXorZ%2FGUbYfV3RDlZwD7fiKKiNf0%2Fimg.jpg)

시계열 데이터의 오차항은 횡단면 자료의 오차항과 다르다. 횡단면 자료의 오차항에는 i.i.d 가정을 적용할 수 있었지만, 시계열 그럴 수 없다.

특히 독립(independent)을 가장 만족하기 어렵다. 간단하게 어제의 주가가 1,000원이라면 오늘의 주가는 1,000원 근처의 값에서 큰 변화를 보이지 않을 것이다.

따라서 어제의 관측치가 오늘의 관측치 값에 영향을 주는 자료 간 자기상관(Autocorrelation)이 존재하기 떄문에 이를 고려한 분석법이 필요하다.

![img](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FkYOVg%2FbtqM9q6dIf8%2FeAQFg3PklLTUyGYSx2uqKK%2Fimg.png)

만약 위와 같이 잔차 간 어떠한 패턴이 보일 경우, 잔차 간 자기상관이 존재한다고 생각할 수 있다. 보통은 양의 상관인 경우가 많고 음인 경우는 많지 않다.

Durbin-Watson Test는 [ 1 - 시차 전 잔차 ]
가 통계적으로 유의한지 파악하는 기법이다.

![img](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FrrmLU%2FbtqM643IJo1%2F3w30yhrpdRkUMGjeZxvTvK%2Fimg.png)

AR(1) 모형은 t시점에서의 관측값을 반응변수로 하고, (t-1)시점에서의 관츠값을 설명변수로 갖는 모형으로 자기 자신의 과거에 회귀시긴다는 의미에서 AR(AutoRegressionn)이라는 표현을 사용한다.


![img](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2Foi9VN%2FbtqM8PysgvD%2Fb5mHNWOTneDUuvNuDJqDf0%2Fimg.png)

$$p{̂} = 0$$ 
일때 d-통계량의 값이 2에 가까우면 자기상관이 없음

$$p{̂} > 0$$ 
일때 d-통계량의 값이 0과 2사이에 있다면 양의 자기상관

$$p{̂} < 0$$ 
일때 d-통계량의 값이 2와 4 사이에 있으면 음의 자기상관을 갖는다.

![img](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2Fc8IokG%2FbtqNe6SNaDR%2FC9BdUDetNEDFkKQb4ol1kk%2Fimg.png)

이때 하한값과 상한값은 유의수준, 관측치 수, 설명변수 개수에 따라 결정된다.
