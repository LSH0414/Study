# 판단 예측(judgmental forecasting)

판단예측은 실제로 사용하는 경우가 매우 흔하다. 대표적인 상황은 아래와 같다.
- 이용할 수 있는 데이터가 없어서, 통계 기법을 적용할 수 없고 할 수 있는 접근 방식이 판단 예측밖에 없는 경우
- 이용할 수 있는 데이터가 있고, 통계적인 예측값을 생성했고, 그리고 나서 판단을 이용하여 조정하는 경우
- 이용할 수 있는 데이터가 있고, 통계적인 예측값과 판단 예측값을 독립적으로 내고 그 둘을 결합하는 경우.

## - 한계점
판단 예측의 가장 큰 한계점은 예측값이 주관적이라는 점이다. 그렇기에 수학적인 식에서 나오는 통계적인 예측값과 달리 일관적이지 않다.

</br>

## - 핵심 원리
판단 예측의 한계점을 극복하기 위해서는 체계적으로 잘 짜여진 접근 방식을 사용해야한다.

- 예측 작업을 간결하고 분명하게
> 예측 작업을 표현할때는 무슨 작업인지 모두가 분명하게 아는 것이 중요하다. 용어 정의는 명확하고 포괄적이어야 한다. 또한, 예측하는 사람이 집중하지 못하게 만들 수 있는 감정을 자극하는 용어나 상관없는 정보는 피하는 것이 중요하다.

- 체계적인 접근 방식 구현
> 예측 작업과 관련 있는 정보를 분류하는 검사표 등의 체계적인 판단 예측 방식을 사용하여 정확도와 일관성을 높일 수 있다.

- 기록, 정당화
> 체계적인 접근 방식에서 실행한 판단 규칙과 가정을 형식화하고 문서 형태로 만드는 것은 규칙을 반복적으로 실행할 수 있어 일관성을 높일 수 있다. 

- 예측값을 체계저으로 평가
> 예측 과정을 체계적으로 추적하고 관찰하는 것을 통해 불규칙한 점을 찾아낼 수도 있다. 특히 예측값을 꾸준히 기록하고, 관련된 고나측값을 사용할 수 있을 때, 피드백 목적으로 기록한 것을 사용하자. 피드백과 평가는 학습과 예측 정확도를 높이는데 도움을 준다.

- 예측가와 사용자를 구분
> 예측가와 사용자는 분명하게 구분되어야 한다. 예를 들어 회사는 재정 목표치에 맞추어 기대하는 판매량이 있을 것이다. 이때 예측값과 수량이 많이 다를 수 있다. 이런 상황에서 단순히 예측값을 제공하는게 아니라 사용자와 꾸준한 커뮤니케이션을 통해 예측값까지 도달하는 과정을 이해시키는 것이 중요하다. 이는 사용자로부터 신뢰를 얻을 수 있는 방법이다. 사용자가 목표치를 정한다면 예측가는 예측값을 내는것이다. 이 두 가지는 다른 명확히 구분되어야 한다.

</br></br>

## - 예측 기법

### 델파이 기법
델파이 기법은 1950년대 랜드의 올라프 헬머오와 노르만 댈키가 특정한 군사 문제를 두리기 위해 개발한 기법이다. **집단이 낸 예측값이 개인이 낸 예측값보다 더 정확하다**는 것이 핵심 가정이다.

#### 델파이 기법의 5단계
1. 전문가를 모은다.
2. 예측 작업과 과제를 정하고 각 전문가에게 분배
3. 각 전문가에게 초기 예측값과 그에 대한 이유를 받아 요약한다.
4. 각 전문가에게 요약된 피드백을 제공하고 다시 예측하고 이유를 받는다.(만족스러운 수준으로 합의가 이뤄질 때까지 반복)
5. 각 전문가의 예측값을 모아 최종 예측값을 구성

#### 한계점
시간 낭비일 수 있다. 시간이 길어지면 참여하는 사람의 관심도나 결집력이 떨어질 수 있다.

</br>

### 유사점 예측
실제로 종종 사용되는 유용한 판단 접근 방식 중 하나이다. 간단한 예로 감정 과정으로 주택 가격을 결정하는 것이다. 감정 과정에서는 해당 지역에서 판매된 비슷한 특징이 있는 집과 비교하여 주택의 시장 가격을 추정한다. 비슷한 정도는 고려하는 속성에 따라 다르다.

#### 한계점
유사점 예측은 한 가지 유사점 비교보다는 여러가지 유사점 비교에 근거한 예측값을 사용해야한다. 그로 인한 편향이 생길 수 있다. 그런데 이런 편향을 찾기가 수비지 않다. 그렇기에 체계적인 접근 방식을 사용하여 해결하는것이 효과적이다.

</br>

### 시나리오 예측
시나리오 예측은 모든 가능한 요인과 동인들의 상대적인 영향, 상호작용, 예측할 목표 등을 고려하여 시나리오를 만든다. 그렇기에 넓은 범위의 예측값이 나오고 극단적인 경우가 포함된다.

</br></br>

## - 판단 조정
판단 조정은 과거 데이터를 사용할 수 있고 통계적인 예측값을 내는데 과거 데이터를 사용하는 상황에서 할 수 있다. 판단 조정의 간단한 예로 판매량 예측에 프로모션, 공휴일, 최근 이벤트 같이 통계 모델에서 고려되지 않은 요소를 통합하는 것이다. 하지만 이런 판단 조정에는 편향과 한계점이 따르고 이를 최소하하는 전략을 구사해야한다.

### 조정은 조심스럽게
통계 모델은 데이터의 패턴을 고려할 때 훨씬 좋은 성능을 발휘한다. 판단 조정을 잘못 고려한다면 정화도가 떨어질 가능성이 있다. 따라서 판단 조정은 유의미한 추가 정보를 가지고 있거나 조정이 필요하다는 강력한 증거가 있을때 사용하는 것이 합리적이다.
