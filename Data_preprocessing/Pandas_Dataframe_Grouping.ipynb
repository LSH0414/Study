```python
import pandas as pd
import numpy as np
```

```python
path = '/Users/seokholee/Downloads/trade_history.csv의 사본'
```

```python
df  = pd.read_csv(path, index_col=0)
```

```python
df.head()
```

```python
target = df[df['streamer']=='한동숙']
```

```python
target.groupby(['streamer','spid','grade'])[['trade_type', 'value']].sum()
```

```python
target.groupby(['streamer','spid', 'grade','trade_type'])[['value']].sum()
```

```python
target.groupby(['streamer','spid', 'grade','trade_type'])[['value']].sum().unstack() # unstack level옵션으로 원하는 index를 col로 이동시킬 수 있음
```

```python
target.pivot_table(index=['streamer','spid','grade'], columns = ['trade_type'], values=['value'], aggfunc='sum')
```
